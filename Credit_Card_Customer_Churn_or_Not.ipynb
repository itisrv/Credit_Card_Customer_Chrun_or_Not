#Visualizing Numerical Data

```python
import numpy as np
def num_summary(data,x):
    x_min=data[x].min()
    x_max=data[x].max()
    Q1=data[x].quantile(0.25)
    Q2=data[x].quantile(0.5)
    Q3=data[x].quantile(0.75)


    data_dict={"Min":x_min,"Q1":Q1,"Q2":Q2,"Q3":Q3,"Max":x_max}
    df=pd.DataFrame(data=data_dict,index=["values"])
    print('-'*90)
    print(df)
    print('-'*90)
    # Set the figure size
    plt.figure(figsize=(16, 8))

    # Adjust vertical space between subplots
    plt.subplots_adjust(hspace=0.6)

    # Set color palette
    sns.set_palette("Pastel1")

    # Plotting subplots
    # Subplot 1: Density Distribution with mean and median lines
    plt.subplot(221)
    sns.distplot(data[x])
    plt.axvline(np.mean(data[x]), color="red", linestyle="--")  # Mean
    plt.axvline(np.median(data[x]), color="orange", linestyle="-")   # Median
    plt.title(f"{x.capitalize()} Density Distribution")


    # Subplot 3: Boxplot
    plt.subplot(222)
    sns.boxplot(x=data[x], showmeans=True)
    plt.title(f"{x.capitalize()} Boxplot")


    plt.show()
```

##Data preprocessing before training

```python

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit_transform(df.drop(['attrition_flag'],axis=1))
scaled_features = scaler.transform(df.drop('attrition_flag',axis=1))
scaled_features = pd.DataFrame(scaled_features, columns = df.columns[:-1])
scaled_features.head()
```

```python
X = scaled_features
y = df['attrition_flag']
```

```python
#splitting into train and test sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
x_train.shape,y_train.shape
```

```python
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier,GradientBoostingClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

##Oversampling the dataset as it is imbalanced

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy = 1.0)

x_train_after_sampling, y_train_after_sampling = sm.fit_resample(x_train, y_train)
```

```python
print(f"Before Sampling (1):{(y_train==1).sum()}")
print(f"Before Sampling (0):{(y_train==0).sum()}")
print(f"After Sampling (1):{(y_train_after_sampling==1).sum()}")
print(f"After Sampling (0):{(y_train_after_sampling==0).sum()}")
```

```python
dtc=DecisionTreeClassifier()
rf=RandomForestClassifier()
ada=AdaBoostClassifier()
xgb=GradientBoostingClassifier()
#accuracy_scores=[]

models=[dtc,rf,ada,xgb]
for model in models:
    model.fit(x_train_after_sampling,y_train_after_sampling)
    y_pred=model.predict(x_test)
    print(f"{model}")
    print('-'*90)
    print(classification_report(y_test, y_pred))
    print('-'*90)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python

param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [2, 4],
    'subsample': [0.8, 1.0]
}
grid_search = GridSearchCV(estimator=xgb, param_grid=param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)

# Fit GridSearchCV to the data
grid_search.fit(x_train_after_sampling, y_train_after_sampling)

# Print best parameters and best score
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation accuracy: {:.2f}".format(grid_search.best_score_))


best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)
test_accuracy = accuracy_score(y_test, y_pred)
print("Test accuracy with best model: {:.2f}".format(test_accuracy))
```
